```python


import pandas as pd
import numpy as np

# -------------------------------------------------------
# STEP 1: Load / Read the dataset
# -------------------------------------------------------
np.random.seed(42)

data = {
    "transaction_id": range(1, 21),
    "customer_id": np.random.randint(100, 110, 20),
    "product": np.random.choice(["Laptop", "Mobile", "Headphones", "Tablet"], 20),
    "quantity": np.random.randint(1, 5, 20),
    "unit_price": np.random.choice([500, 800, 1200, 25000, 45000], 20),
    "discount_pct": np.random.choice([0, 5, 10, 15], 20),
    "region": np.random.choice(["North", "South", "East", "West"], 20),
}

df = pd.DataFrame(data)
print("Raw Data Sample:")
print(df.head(), "\n")

# -------------------------------------------------------
# STEP 2: Apply business rules and transformations
# -------------------------------------------------------

# Business Rule 1: Calculate gross amount
df["gross_amount"] = df["quantity"] * df["unit_price"]

# Business Rule 2: Apply discount to get net amount
df["net_amount"] = df["gross_amount"] * (1 - df["discount_pct"] / 100)

# Business Rule 3: Flag high-value transactions (> 50,000)
df["high_value_flag"] = df["net_amount"] > 50000

# Business Rule 4: Categorize order size
def categorize_order(qty):
    if qty <= 1:
        return "Small"
    elif qty <= 3:
        return "Medium"
    else:
        return "Large"

df["order_size"] = df["quantity"].apply(categorize_order)

print("Processed Data Sample:")
print(df.head(), "\n")

# -------------------------------------------------------
# STEP 3: Generate meaningful outputs
# -------------------------------------------------------

# Output 1: Total revenue by region
revenue_by_region = df.groupby("region")["net_amount"].sum().sort_values(ascending=False)
print("Total Revenue by Region:")
print(revenue_by_region, "\n")

# Output 2: Best-selling product by quantity
top_products = df.groupby("product")["quantity"].sum().sort_values(ascending=False)
print("Units Sold by Product:")
print(top_products, "\n")

# Output 3: High value transactions
high_value_txns = df[df["high_value_flag"]]
print(f"Number of High-Value Transactions (>50,000): {len(high_value_txns)}\n")

# Output 4: Overall summary
summary = {
    "Total Transactions": len(df),
    "Total Revenue": round(df["net_amount"].sum(), 2),
    "Average Order Value": round(df["net_amount"].mean(), 2),
    "Total Discount Given": round((df["gross_amount"] - df["net_amount"]).sum(), 2),
}
print("Business Summary:")
for k, v in summary.items():
    print(f"  {k}: {v}")

# -------------------------------------------------------
# STEP 4: Save processed output
# -------------------------------------------------------
df.to_csv("processed_sales_data.csv", index=False)
print("\nProcessed data saved to 'processed_sales_data.csv'")
```
